```python
import numpy as np 
import matplotlib.pyplot as plt 
import glob
import soundfile as sf
import librosa as lb
import acoustics
from scipy.optimize import curve_fit
from scipy.signal import hilbert, fftconvolve
from IPython.display import Audio
eps = np.finfo(float).eps
reales_path = '/mnt/datasets/impulsos/reales/'
sinteticos_path = '/mnt/datasets/impulsos/sinteticos/audios'
lista_rir_reales = glob.glob(reales_path+'**/*.wav',recursive=True)
lista_rir_sintetic = glob.glob(sinteticos_path+'**/*.wav',recursive=True)

def load_rir(path, target_fs, norm=True):
    rir, fs = lb.load(path, sr = target_fs)
    if norm:
        rir = rir / np.max(abs(rir))
    return rir, fs

def temporal_decompose(rir, fs, win = 0.0025):
    t_d = np.argmax(rir) # direct path
    t_o = int((win) * fs) #tolerance window in samples (2.5 ms)
    early= rir[(t_d - t_o):(t_d + t_o)+1]
    late = rir[t_d + t_o+1:]
    return early, late

def normalize(arr, valor=None):
    if valor is not None:
        return arr/np.max(abs(valor))
    return arr/np.max(abs(arr))

def promediar_intervalos(señal, intervalo):
    frames_enteros = int(len(señal)/intervalo)
    resto = len(señal)%intervalo
    salida = np.empty(len(señal))
    for n_frame in range(frames_enteros):
        salida[n_frame*intervalo:(n_frame+1)*intervalo] = señal[n_frame*intervalo:(n_frame+1)*intervalo].mean()
    salida[len(señal)-resto:] = señal[len(señal)-resto:].mean() 
    return salida
```

# Proceso de determinación de piso de ruido de Lundeby et al. 

PASO 1: El valor absoluto de la respuesta al impulso es promediado en intervalos de tiempo locales de $10-50 ms$ para obtener una curva suavizada sin perder las pendientes cortas. 

```python
#cargo una rir y me quedo con la parte late
rir, fs = load_rir(lista_rir_reales[17], 16000)
rir_early, rir_late = temporal_decompose(rir, fs)
rir_late = normalize(abs(rir_late))
integracion = 30 #milisegundos
intervalo_temporal = int((integracion * 0.001) * fs)
#env = fftconvolve(rir_late, np.ones(intervalo_temporal)/intervalo_temporal)[intervalo_temporal-1:] #filtro de media movil 
env = promediar_intervalos(rir_late, intervalo_temporal)
```

```python
#Normalizo segun la envolvente 
env = normalize(env) #No afecta al proceso? si trabajo solo sobre env en cuanto a vaores en Y. revisar

#Escala logaritmica
env_db = 20*np.log10(env+eps)
rir_late_db= 20*np.log10((rir_late)+eps)
t = np.linspace(0,len(rir_late)/fs,len(rir_late)) #vector de tiempo

fig, (ax1,ax2) = plt.subplots(2)
ax1.plot(t, rir_late, label='Impulso al cuadrado')
ax1.plot(t, env, label = 'Curva suavizada por promedios')
ax1.legend(fontsize=18)

ax2.plot(rir_late_db, label='Impulso al cuadrado')
ax2.plot(env_db, label = 'Curva suavizada por promedios')
ax2.legend(fontsize=18)
#ax2.set_xlim([0,2])
ax2.grid()

plt.subplots_adjust(right=2, top=2)
```

PASO 2: Se hace una primera estimación del ruido de fondo utilizando un segmento de tiempo que contenga el último 10% de la respuesta al impulso. Esto representa una selección temporal estadisticamente razonable sin un error sistematico elevado si el decay continua al final de la respuesta impulsiva. 

```python
ventana_ruido = int(len(env)*0.1) #10% de la señal
nivel_ruido_db = env_db[-ventana_ruido:].mean()

distancia_al_piso = 8 #dB
umbral = nivel_ruido_db + distancia_al_piso

x_umbral = np.where(env_db<umbral)[0][0]
y_umbral = env_db[x_umbral]

plt.scatter(x_umbral, y_umbral, s=1000, marker='x', c='red') 
plt.plot(rir_late_db, label='Impulso al cuadrado')
plt.plot(env_db, label = 'Curva suavizada por promedios')
plt.legend(fontsize=18)
#plt.xlim([28682, 31682])
#lt.ylim([-60,0])
plt.grid()
plt.subplots_adjust(right=2, top=2)
```

PASO 3: La pendiente de caida se estima usando una regresión lineal entre el punto temporal de 0dB y el primer intervalo de entre 5 y 10 dB por encima del piso de ruido

```python
x_0dB = 0
y_0dB = env_db[0]

A = np.vstack([[x_0dB, x_umbral], [1, 1]]).T
y = [y_0dB, y_umbral]
m, c = np.linalg.lstsq(A, y, rcond=None)[0]

def curva(x, m, c):
    return m * x + c

regresion = curva(np.linspace(0, len(env)-1, len(env)), m, c)

plt.scatter(x_umbral, y_umbral, s=500, marker='x', c='red', label='5dB sobre el ruido de fondo') 
plt.plot(rir_late_db, label='Impulso al cuadrado')
plt.plot(env_db, label = 'Curva suavizada por promedios')
plt.plot(regresion)
plt.legend(fontsize=18)
#plt.xlim([30000, 32000])
#plt.ylim([-60,0])
plt.grid()
plt.subplots_adjust(right=2, top=2)
```

PASO 4: Se determina un punto de cruce preliminar en el cruce entre la pendiente estimada y el nivel de ruido de fondo

```python
cruce_index = int((nivel_ruido_db-c)/m)
cruce_valor = curva(cruce_index, m, c)

plt.scatter(cruce_index, cruce_valor, s=500, marker='x', c='green', label='Piso de ruido') 
plt.scatter(x_umbral, y_umbral, s=500, marker='x', c='red', label='5dB sobre el ruido de fondo') 
plt.plot(rir_late_db, label='Impulso al cuadrado')
plt.plot(env_db, label = 'Curva suavizada por promedios')
plt.plot(regresion)
plt.legend(fontsize=18)
#plt.xlim([30000, 32000])
plt.ylim([-60,0])
plt.grid()
plt.subplots_adjust(right=2, top=2)
```

PASO 5: Se calcula un nuevo largo de intervalo de acuerdo a la pendiente obtenida, de manera de formar entre 3 y 10 intervalos por de 10 dB de caida

```python
nuevo_intervalo = int((-2-c)/m)
```

PASO 6: El valor absoluto del impulso es promediado nuevamente en los nuevos intervalos locales.

```python
#env_nueva = fftconvolve(rir_late, np.ones(nuevo_intervalo)/nuevo_intervalo)[nuevo_intervalo-1:]
env_nueva = promediar_intervalos(rir_late, nuevo_intervalo)
env_nueva = normalize(env_nueva)
env_nueva_db = 20*np.log10(env_nueva+eps)
```

```python
plt.scatter(cruce_index, cruce_valor, s=500, marker='x', c='green', label='Piso de ruido') 
plt.scatter(x_umbral, y_umbral, s=500, marker='x', c='red', label='5dB sobre el ruido de fondo') 
plt.plot(rir_late_db, label='Impulso al cuadrado')
plt.plot(env_nueva_db, label = 'Nueva envolvente')
plt.plot(env_db, label = 'Envolvente vieja')
plt.plot(regresion)
plt.legend(fontsize=18)
#plt.xlim([30000, 32000])
#plt.ylim([-60,1])
plt.grid()
plt.subplots_adjust(right=2, top=2)
```

PASO 7: Se determina nuevamente el nivel de piso de ruido. El segmento de ruido a evaluar debe comenzar desde un punto correspondiente a 5-10 dB de caida luego del punto de cruce o un minimo del 10% de la señal

```python
index_inicial = np.where(env_nueva_db < cruce_valor-5)[0]
if index_inicial.size == 0: 
    nuevo_nivel_ruido_db = env_nueva_db[-ventana_ruido:].mean()
    print('Nuevo cruce fuera del rango')
elif len(env_nueva_db[index_inicial[0]:])<ventana_ruido:
    nuevo_nivel_ruido_db = env_nueva_db[-ventana_ruido:].mean()
    print('No hay 10% de señal luego del nuevo umbral')
else:
    print('Se calcula promediando todo lo que esta despues del nuevo cruce')
    nuevo_nivel_ruido_db = env_nueva_db[index_inicial[0]:].mean()
        
distancia_al_piso = 10 #dB
umbral = nuevo_nivel_ruido_db + distancia_al_piso

x_umbral = np.where(env_nueva_db<umbral)[0][0]
y_umbral = env_nueva_db[x_umbral]

plt.scatter(x_umbral, y_umbral, s=1000, marker='x', c='red', label='Nuevo umbral') 
plt.plot(rir_late_db, label='Impulso al cuadrado')
plt.plot(env_nueva_db, label = 'Curva suavizada por promedios')
plt.legend(fontsize=18)
#plt.xlim([28682, 31682])
plt.ylim([-60,0])
plt.grid()
plt.subplots_adjust(right=2, top=2)
```
