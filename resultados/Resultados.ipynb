```python
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
import glob
import seaborn
```

```python
def plot_results(PESOS, TEST):
    
    path = 'valores/'+PESOS+TEST+'*.npy'

    valores = glob.glob(path)
    val_dict = {}
    for val in valores:
        arr = np.load(val)
        label = val.split('/')[-1].split('.')[0]
        val_dict[label] = arr
    dereverb = [np.median(val_dict['SDR_dereverb']), np.median(val_dict['SRMR_dereverb']), np.median(val_dict['ESTOI_dereverb'])]
    reverb = [np.median(val_dict['SDR_reverb']), np.median(val_dict['SRMR_reverb']), np.median(val_dict['ESTOI_reverb'])]
    
    print('SDR reverb: {:.2f} \tSDR dereverb: {:.2f}'.format(np.median(val_dict['SDR_reverb']), np.median(val_dict['SDR_dereverb'])))
    print('SRMR reverb: {:.2f} \tSRMR dereverb: {:.2f}'.format(np.median(val_dict['SRMR_reverb']), np.median(val_dict['SRMR_dereverb'])))
    print('ESTOI reverb: {:.2f} \tESTOI dereverb: {:.2f}'.format(np.median(val_dict['ESTOI_reverb']), np.median(val_dict['ESTOI_dereverb'])))

    """
    plt.figure(1, figsize=(15,5))
    plt.subplot(1, 3, 1)
    plt.hist(val_dict['SDR_reverb'], label = 'SDR_reverb', alpha=0.8 )
    plt.hist(val_dict['SDR_dereverb'], label = 'SDR_dereverb', alpha=0.8 )
    plt.legend()

    plt.subplot(1, 3, 2)
    plt.hist(val_dict['SRMR_reverb'], label = 'SRMR_reverb', alpha=0.8 )
    plt.hist(val_dict['SRMR_dereverb'], label = 'SRMR_dereverb', alpha=0.8 )
    plt.legend()

    plt.subplot(1, 3, 3)
    plt.hist(val_dict['ESTOI_reverb'], label = 'ESTOI_reverb', alpha=0.8 )
    plt.hist(val_dict['ESTOI_dereverb'], label = 'ESTOI_dereverb', alpha=0.8 )
    plt.legend()

    plt.show()
    """
    return dereverb, reverb 
```

# Entrenado con reverb reales

```python
SDR = {'Real' : [], 'Generada' : [], 'Aumentada' : []} # Entrenamiento
SRMR = {'Real' : [], 'Generada' : [], 'Aumentada' : []}
ESTOI = {'Real' : [], 'Generada' : [], 'Aumentada' : []}
REVERB1 = {'SDR' : [], 'SRMR' : [], 'ESTOI' : []}
REVERB2 = {'SDR' : [], 'SRMR' : [], 'ESTOI' : []}
REVERB3 = {'SDR' : [], 'SRMR' : [], 'ESTOI' : []}
```

# RESULTADOS PARCIALES

```python
# Referencias a reverb
a = pd.DataFrame.from_dict(REVERB1, orient = 'index', columns=['Real', 'Generada', 'Aumentada'])
b = pd.DataFrame.from_dict(REVERB2, orient = 'index', columns=['Real', 'Generada', 'Aumentada'])
c = pd.DataFrame.from_dict(REVERB3, orient = 'index', columns=['Real', 'Generada', 'Aumentada'])
resultados_reverb = (a + b + c) / 3
resultados_reverb
```

```python
resultados_sdr = pd.DataFrame.from_dict(SDR, orient = 'index', columns=['Real', 'Generada', 'Aumentada'])
resultados_sdr['Real'] = resultados_sdr['Real'] - resultados_reverb['Real']['SDR']
resultados_sdr['Generada'] = resultados_sdr['Generada'] - resultados_reverb['Generada']['SDR']
resultados_sdr['Aumentada'] = resultados_sdr['Aumentada'] - resultados_reverb['Aumentada']['SDR']
```

```python
resultados_srmr = pd.DataFrame.from_dict(SRMR, orient = 'index', columns=['Real', 'Generada', 'Aumentada'])
resultados_srmr['Real'] = resultados_srmr['Real'] - resultados_reverb['Real']['SRMR']
resultados_srmr['Generada'] = resultados_srmr['Generada'] - resultados_reverb['Generada']['SRMR']
resultados_srmr['Aumentada'] = resultados_srmr['Aumentada'] - resultados_reverb['Aumentada']['SRMR']
```

```python
resultados_estoi = pd.DataFrame.from_dict(ESTOI, orient = 'index', columns=['Real', 'Generada', 'Aumentada'])
resultados_estoi['Real'] = resultados_estoi['Real'] - resultados_reverb['Real']['ESTOI']
resultados_estoi['Generada'] = resultados_estoi['Generada'] - resultados_reverb['Generada']['ESTOI']
resultados_estoi['Aumentada'] = resultados_estoi['Aumentada'] - resultados_reverb['Aumentada']['ESTOI']
```

```python
ylabel = ['Reales', 'Generadas', 'Aumentadas']
xlabel = ['Reales', 'Generadas', 'Aumentadas']
seaborn.set(font_scale=1.4)


plt.title('\u0394 SDR', fontsize=18)
seaborn.heatmap(resultados_sdr, annot=True, cmap='Blues', yticklabels=False)
plt.ylabel('Conjunto de Entrenamiento',fontsize=16)
plt.xlabel('Conjunto de Evaluaci√≥n', fontsize=16)
plt.yticks([0.5,1.5,2.5], ylabel, ha = 'center', size=13)
plt.xticks([0.5,1.5,2.5], xlabel, size=13)
plt.tick_params(axis='y', which='major', pad=42)
plt.savefig('imagenes/prueba1_SDR.png',bbox_inches='tight')
plt.show()
```

```python
plt.title('\u0394 SRMR', fontsize=18)
seaborn.heatmap(resultados_srmr, annot=True, cmap='Blues', yticklabels=False)
plt.ylabel('Conjunto de Entrenamiento',fontsize=16)
plt.xlabel('Conjunto de Evaluaci√≥n', fontsize=16)
plt.yticks([0.5,1.5,2.5], ylabel, ha = 'center', size=13)
plt.xticks([0.5,1.5,2.5], xlabel, size=13)
plt.tick_params(axis='y', which='major', pad=42)
plt.savefig('imagenes/prueba1_SRMR.png',bbox_inches='tight')
plt.show()
```

```python
plt.title('\u0394 ESTOI', fontsize=18)
seaborn.heatmap(resultados_estoi, annot=True, cmap='Blues', yticklabels=False)
plt.ylabel('Conjunto de Entrenamiento',fontsize=16)
plt.xlabel('Conjunto de Evaluaci√≥n', fontsize=16)
plt.yticks([0.5,1.5,2.5], ylabel, ha = 'center', size=13)
plt.xticks([0.5,1.5,2.5], xlabel, size=13)
plt.tick_params(axis='y', which='major', pad=42)
plt.savefig('imagenes/prueba1_ESTOI.png',bbox_inches='tight')
plt.show()
```
