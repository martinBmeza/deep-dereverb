```python
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
import pandas as pd
import glob
import seaborn
import ptitprince as pt
```

```python
def reject_outliers(data, m = 2.0):
    d = np.abs(data - np.median(data))
    mdev = np.median(d)
    s = d/mdev if mdev else 0.
    return data[s<m]
```

# SDR

```python
#SDR
ordenado = np.load('valores/ordenado/SDR_dereverb.npy')
desordenado = np.load('valores/inverso/SDR_dereverb.npy')
random = np.load('valores/random/SDR_dereverb.npy')

#ordenado = reject_outliers(ordenado)
#desordenado = reject_outliers(desordenado)
#random = reject_outliers(random)
```

```python
label = ['TR Creciente' for i in range(len(ordenado))]
data_ordenado = pd.DataFrame(list(zip(label, ordenado)), columns = [' ', 'SDR'])

label = ['TR Decreciente' for i in range(len(desordenado))]
data_desordenado = pd.DataFrame(list(zip(label, desordenado)), columns = [' ', 'SDR'])

label = ['TR Aleatorio' for i in range(len(random))]
data_random = pd.DataFrame(list(zip(label, random)), columns = [' ', 'SDR'])

data = pd.concat([data_ordenado, data_desordenado, data_random], ignore_index=True)
```

```python
from matplotlib import rcParams
rcParams['text.usetex'] = True

dx=' '; dy="SDR"; ort="v"; pal = "Set2"

pt.RainCloud(x = dx, y = dy, data = data, palette = pal, width_viol = .5, orient = ort, bw=.8, rain_alpha=0.1)
plt.grid()
plt.savefig('imagenes/sdr.png', bbox_inches='tight')
print('=======MEDIANAS=======')
print('TR Creciente: ', np.median(ordenado))
print('TR Decreciente: ', np.median(desordenado))
print('TR Aleatorio: ', np.median(random))
```
