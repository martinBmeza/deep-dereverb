```python
#!pip install rir-generator
import tqdm
import numpy as np
import soundfile as sf
import rir_generator as rirgen
from generacion_rir import *
```

# ROOM 1

```python
# Parametros de generacion
room1_dim = [3,4,5]
source = [1,3,1.5]
receiver_positions = rand_positions(room1_dim, 15)
trs = [0.5, 0.75, 1.0]
save_path = '/home/martin/deep-dereverb/data/rir_generacion/generados/'
```

```python
plot_room(room1_dim, source, receiver_positions, 'Recinto 1')
```

```python
room1_params = {'c' : 340,
                'fs' : 16000,
                'r' : receiver_positions,
                's' : source,
                'L' : room1_dim,
                'nsample' : 32768}
```

```python
impulsos = []
for tr in tqdm.tqdm(trs):
    h = rirgen.generate(reverberation_time=tr, **room1_params)
    impulsos.append(h)
```

```python
# Guardando los archivos generados
for i,h in enumerate(impulsos):
    tr_label = str(trs[i])
    for receiver in range(h.shape[-1]):
        filename = save_path + tr_label + 'p' + str(receiver) + '.wav'
        impulso = norm_to_save(h[:,receiver])
        sf.write(filename, impulso, 16000)
```
