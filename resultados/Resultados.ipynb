```python
import matplotlib.pyplot as plt 
import numpy as np
import glob
```

```python
def plot_results(PESOS, TEST):
    
    path = 'valores/'+PESOS+TEST+'*.npy'

    valores = glob.glob(path)
    val_dict = {}
    for val in valores:
        arr = np.load(val)
        label = val.split('/')[-1].split('.')[0]
        val_dict[label] = arr

    plt.figure(1, figsize=(15,5))
    plt.subplot(1, 3, 1)
    plt.hist(val_dict['SDR_reverb'], label = 'SDR_reverb', alpha=0.8 )
    plt.hist(val_dict['SDR_dereverb'], label = 'SDR_dereverb', alpha=0.8 )
    plt.legend()
    print('SDR reverb: {:.2f} \tSDR dereverb: {:.2f}'.format(np.median(val_dict['SDR_reverb']), np.median(val_dict['SDR_dereverb'])))

    plt.subplot(1, 3, 2)
    plt.hist(val_dict['SRMR_reverb'], label = 'SRMR_reverb', alpha=0.8 )
    plt.hist(val_dict['SRMR_dereverb'], label = 'SRMR_dereverb', alpha=0.8 )
    plt.legend()
    print('SRMR reverb: {:.2f} \tSRMR dereverb: {:.2f}'.format(np.median(val_dict['SRMR_reverb']), np.median(val_dict['SRMR_dereverb'])))

    plt.subplot(1, 3, 3)
    plt.hist(val_dict['ESTOI_reverb'], label = 'ESTOI_reverb', alpha=0.8 )
    plt.hist(val_dict['ESTOI_dereverb'], label = 'ESTOI_dereverb', alpha=0.8 )
    plt.legend()
    print('ESTOI reverb: {:.2f} \tESTOI dereverb: {:.2f}'.format(np.median(val_dict['ESTOI_reverb']), np.median(val_dict['ESTOI_dereverb'])))

    plt.show()
```

# Entrenado con reverb reales

```python
PESOS = 'pesos_reales/'
TEST = 'real/'
print('Testeado en reverb reales')
plot_results(PESOS, TEST)
```

```python
print('Testeado en reverb generadas')
TEST = 'gen/'
plot_results(PESOS, TEST)
```

```python
print('Testeado en reverb aumentadas')
TEST = 'aug/'
plot_results(PESOS, TEST)
```
