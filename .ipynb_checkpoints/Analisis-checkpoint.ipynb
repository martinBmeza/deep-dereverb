# Visualizacion de rango dinámico
Se recorre el dataset para ver cuales son los rangos dinamicos en los que se encuentran las diferentes transformadas. Esto sera util para tomar un criterio de normalizacion a la hora de entrenar el modelo

```python
import random
def PrePipeline(speech_path, rir_path):
    
    speech, speech_fs = librosa.load(speech_path, sr=16000)                                                                                                                 
    rir, rir_fs = librosa.load(rir_path, sr=16000)                                                                                                                         

    # Elimino el delay del impulso                                                                                                                                  
    delay_shift = np.argmax(rir)                                                                                                                                                                                                                                                                             
    rir = rir[delay_shift:]

    #Normalizo el impulso
    rir = rir / np.max(abs(rir))

    #Convoluciono. Obtengo audio con reverb                                                                                                                           
    reverb = fftconvolve(speech, rir)                                                                                                                                 

    #Padeo el audio anecoico. Obtengo el audio clean                                                                                                                  
    clean = np.pad(speech,(0,len(rir)-1), 'constant', constant_values=(eps,eps))                                                              

    #PRUEBA
    reverb = reverb / abs(reverb).max()
    clean = clean / abs(clean).max()
    
    
    #genero las STFTs en dB
    stft_clean = librosa.stft(clean, n_fft=512, hop_length=128)# DE ENTRADA YA PUEDO ESTAR DESCARTANDO EL BIN DE FFREQ DE MAS
    spectrogram_clean = np.abs(stft_clean)
    log_spectrogram_clean = librosa.amplitude_to_db(spectrogram_clean)

    stft_reverb = librosa.stft(reverb, n_fft=512, hop_length=128)# DE ENTRADA YA PUEDO ESTAR DESCARTANDO EL BIN DE FFREQ DE MAS
    spectrogram_reverb = np.abs(stft_reverb)
    log_spectrogram_reverb = librosa.amplitude_to_db(spectrogram_reverb)
    return log_spectrogram_reverb.max(), log_spectrogram_reverb.min(), log_spectrogram_clean.max(), log_spectrogram_clean.min()

rev_maxs = []
clean_maxs = []

rev_mins = []
clean_mins = []

import tqdm
for audio in tqdm.tqdm(speech_list):
    [rir] = random.sample(rir_list, 1)
    rev_max, rev_min, clean_max, clean_min = PrePipeline(audio, rir)
    rev_maxs.append(rev_max)
    rev_mins.append(rev_min)
    clean_maxs.append(clean_max)
    clean_mins.append(clean_min)
print(np.min(rev_mins), np.max(rev_maxs))
print(np.min(clean_mins), np.max(clean_maxs))
```

# Pipeline de transformación

Los pasos se aplican en el siguiente orden: 

    - Cargar audio de habla
    - Cargar RIR aleatoria
    - Eliminar delay del RIR
    - Convolucionar para obtener audio reverberado
    - Padear el audio clean para igualar dimensiones
    - Transformar STFT ambos audios
    - Pasar a escala logaritmica
    - Normalizar amplitudes [0,1]
    - Generar mascara 
    
La salida son los pares [espectro_reverberado,espectro_anecoico]

Esta cadena se aplica en dos partes a lo largo del codigo:

Una primera parte estatica, en donde se hace el procesamiento hasta obtener los pares reververados-anecoicos en formato de audios segmentados (paso previo a la transformación STFT). En este punto, los pares de arrays se almacenan en memoria para luego ser usados en el proceso de entrenamiento
    
La segunda parte dinamica se realiza dentro del generador de datos, que a medida que la red necesita batches de datos para entrenarse el generador lee pares de arrays (en forma de audio) y genera los espectros y les aplica los procesos necesarios (extraccion de magnitud, escala logaritimica, normalizacion) y luego entrega estos datos al input de la red. 

```python
array_min, array_max = -55 , 39
 
def normalise(array, range_min, range_max, array_min, array_max):
    norm_array = (array - array_min) / (array_max - array_min)
    norm_array = norm_array * (range_max - range_min) + range_min
    return norm_array

def denormalise(norm_array, array_min, array_max, range_min, range_max):
    array = (norm_array - range_min) / (range_max - range_min)
    array = array * (array_max - array_min) + array_min
    return array

def temporal_decompose(rir, fs, win = 0.0010):
    t_d = np.argmax(rir) # direct path
    t_o = int((win) * fs) #tolerance window in samples
    
    if t_d - t_o < 0:
        start = 0
    else:
        start = t_d - t_o
        
    if t_d + t_o > len(rir) - 1:
        end = len(rir) - 1
    else:
        end = t_d + t_o + 1
    
    early = rir[start:end]
    complete = rir[start:]
    print(len(early))
    return early, complete
```

## Funciones

```python
def predict_model(data, modelo):                                                                                                                                      
  output = [layer.name for layer in modelo.layers]                                                                                                                    
  outputs = []                                                                                                                                                        
  output_names = []                                                                                                                                                   
  inputs = []                                                                                                                                                         
  input_names = []                                                                                                                                                    
  for layer in modelo.layers:                                                                                                                                         
      if hasattr(layer,'is_placeholder'):                                                                                                                             
          inputs.append(layer.output)                                                                                                                                 
          input_names.append(layer.name)                                                                                                                              
      elif layer.name in output:                                                                                                                                      
          outputs.append(layer.output)                                                                                                                                
          output_names.append(layer.name)                                                                                                                             
      else:                                                                                                                                                           
          pass                                                                                                                                                        
  predict_fn = tf.keras.backend.function(inputs = inputs,outputs=outputs)                                                                                             
  activations = predict_fn(data)                                                                                                                                      
  activations = {name: act for name, act in zip(output_names,activations)}                                                                                            
  #print('orden de las entradas: /\n',input_names)                                                                                                                     
  return activations

def preProcessing(reverb, clean):
    
    stft_clean = librosa.stft(clean, n_fft=512, hop_length=128)                                                                                           
    spectrogram_clean = np.abs(stft_clean)                                                                                                                
    log_spectrogram_clean = librosa.amplitude_to_db(spectrogram_clean)                                                                                    
    log_norm_clean = normalise(log_spectrogram_clean, 0, 1, array_min, array_max)                                                                         
    
    stft_reverb = librosa.stft(reverb, n_fft=512, hop_length=128)                                                                                         
    spectrogram_reverb = np.abs(stft_reverb)                                                                                                              
    log_spectrogram_reverb = librosa.amplitude_to_db(spectrogram_reverb)                                                                                  
    log_norm_reverb = normalise(log_spectrogram_reverb, 0, 1, array_min, array_max)
    return log_norm_reverb.reshape(1, 257, 256), log_norm_clean.reshape(1, 257, 256)
```

# Procesando archivos de audio con reverb


```python
import tqdm

def audio_framing(audio, winsize = 32640):                                                                                                                            
                                                                                                                                                                      
    n_frames = int(len(audio)/winsize)                                                                                                                                
    resto = len(audio)%winsize                                                                                                                                        
                                                                                                                                                                      
    if resto == 0:                                                                                                                                                    
        out = np.empty((n_frames, winsize))                                                                                                                           
    else:                                                                                                                                                             
        out = np.empty((n_frames+1, winsize))                                                                                                                         
        out[-1,:resto] = audio[n_frames*winsize:]                                                                                                                     
        out[-1,resto:] = np.zeros(winsize-resto)                                                                                                                      
                                                                                                                                                                      
    for frame in range(n_frames):                                                                                                                                     
        out[frame,:] = audio[frame*winsize:(frame+1)*winsize]                                                                                                         
    audio_pad = np.concatenate((audio, np.zeros(winsize-resto)))
    return out, audio_pad

def predict_audio(audio_path, rir_path, modelo_entrenado):
    '''Hacer predicciones sobre audios sin segmentar'''
    
    # Cargar y segmentar audio de entrada en ventanas que pueda procesar la red 
    audio_in, _ = librosa.load(audio_path, sr=16000)
    rir, _ = librosa.load(rir_path, sr=16000)
    reverb = fftconvolve(audio_in, rir)
    frames_reverb, reverb = audio_framing(reverb)
    
    #Procesando cada frame
    
    predict = []
    for frame_index in tqdm.tqdm(range(frames_reverb.shape[0])):
        
        # Aplicar el preprocesamiento para llegar a las dimensiones de entrada de la red
        frame_pre, _ = preProcessing(frames_reverb[frame_index], frames_reverb[frame_index])

        # pasar cada segmento por la red y recuperar el espectro dereverberado
        dereverb = modelo_entrenado.predict(frame_pre)
        clean_p = denormalise(dereverb.reshape(257, 256), array_min, array_max, 0, 1)
        clean_p = librosa.db_to_amplitude(clean_p)

        # Pasar a audio aplicando griffin lim
        clean_p = librosa.griffinlim(clean_p, n_iter=500, hop_length=128, win_length=512)
        
        # Unir los frames dereverberados para formar el audio de salida
        predict = np.concatenate((predict, clean_p))
        
    return reverb, predict
```
